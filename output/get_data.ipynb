#  GitHubからコードをもらう

APIを叩きます
    https://developer.github.com/v3/
    条件
    キーワード「python」を含むもの
    「most star」順にソートし上位100件
    1度に取ってこられるデータの上限が100のため
上記の条件に当てはまるリポジトリのデータから「html_url」を取り出しCSVに書き出す
CSVファイルのURL一覧をgit clone
取得したリポジトリから「#」を含む「.py」ファイルだけを抽出
Githubは特殊文字を検索できないため一度ローカルに落としてから「#」を検索する方針にしました

```python
import requests
response = requests.get("https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page=1")
```

```python
datas = data['items']
print(type(data["items"]))
```

```python
x = [datas[i]['html_url'] for i in range(len(datas))]
len(x)
```

```python
import csv

with open("urlsets.csv", "a") as csvFile:
  fieldnames = ["url"]
  writer = csv.DictWriter(csvFile, fieldnames=fieldnames)
  writer.writerow({"url":x})
  csvFile.close()
```

```python
import json
import requests
import csv
import time

page = ['https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page=1','https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page=2','https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page=3','https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page=4','https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page=5']

for p in page:
  #url = "https://api.github.com/search/repositories?q=language：Python&sort=stars&order=desc&per_page=100&page="+ p
  print(p)  
  response = requests.get(p)
  data = json.loads(response.text)
  try: 
    x = data["items"]
    y = [x[i]['name'] for i in range(len(x))]
    x = [x[i]['html_url'] for i in range(len(x))]
  except KeyError:
    pass
  #print(x)

  with open("url_Data.txt", "a") as txt:
    for i in range(len(x)):
      txt.write(r'\item ' + y[i] + r' \url{' + x[i] + '} \n')
    time.sleep(3)
```
