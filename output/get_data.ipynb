```python
import json
import requests
import csv
import time

page = ['https://api.github.com/search/repositories?q=languageï¼šPython&sort=stars&order=desc&per_page=100&page=1','https://api.github.com/search/repositories?q=languageï¼šPython&sort=stars&order=desc&per_page=100&page=2','https://api.github.com/search/repositories?q=languageï¼šPython&sort=stars&order=desc&per_page=100&page=3','https://api.github.com/search/repositories?q=languageï¼šPython&sort=stars&order=desc&per_page=100&page=4','https://api.github.com/search/repositories?q=languageï¼šPython&sort=stars&order=desc&per_page=100&page=5']

for p in page:
  #url = "https://api.github.com/search/repositories?q=languageï¼šPython&sort=stars&order=desc&per_page=100&page="+ p
  print(p)  
  response = requests.get(p)
  data = json.loads(response.text)
  try: 
    x = data["items"]
    y = [x[i]['name'] for i in range(len(x))]
    x = [x[i]['html_url'] for i in range(len(x))]
  except KeyError:
    pass
  #print(x)

  with open("url_Data.txt", "a") as txt:
    for i in range(len(x)):
      txt.write(r'\item ' + y[i] + r' \url{' + x[i] + '} \n')
    time.sleep(3)
```
